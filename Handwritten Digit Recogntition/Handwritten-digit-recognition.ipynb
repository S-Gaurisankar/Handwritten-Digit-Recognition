```python
train_df = pd.read_csv("/kaggle/input/handwritten-digit-identification-b1/train.csv",header=None)
test_df = pd.read_csv("/kaggle/input/handwritten-digit-identification-b1/test.csv",header=None)
```

### Dropping column 2 as it has no correlation

```python
train_df.drop([0,1],axis='columns',inplace=True)
```

```python
train_df
```

```python
X = train_df.drop([65],axis="columns")
y = train_df[65]
```

### Normalization is not required as all the data is in range.

## Splitting the data in 70:30 ratio

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,test_size=0.20, random_state=42)
```

# Tuning the model

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(256,128,64,32),random_state=40, max_iter=150, learning_rate_init=0.0013, 
                    activation='relu')
clf.fit(Xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
preds = clf.predict(Xtest)
accuracy = accuracy_score(preds,ytest)
print("Accuracy on validation dataset",accuracy)
```
